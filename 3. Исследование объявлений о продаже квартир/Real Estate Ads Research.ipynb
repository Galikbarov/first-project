Избавимся от пропущенных срок в `floors_total` и `locality_name`, потому что их количество очень мало.

Пропуски в `balcony`, `parks_around3000` и `ponds_around3000` заменим на 0, так как скорее всего их не указали из-за отсутсвия.

Пропуски в `ceiling_height` заменим на медиану.

Пропуски в `is_apartment` заменим на False, как как скорее всего логикая такая же как балконами.

Заменим вещественные переменный на целые, там где это возможно.

```python
df.dropna(subset=['floors_total', 'locality_name'], inplace=True)
df.reset_index(drop=True)

df['balcony'].fillna(0, inplace=True)
df['is_apartment'].fillna(False, inplace=True)
df['parks_around3000'].fillna(0, inplace=True)
df['ponds_around3000'].fillna(0, inplace=True)

ceiling_height_median = df['ceiling_height'].median()
df['ceiling_height'].fillna(ceiling_height_median, inplace=True)

df['balcony'] = df['balcony'].astype('int')
df['floors_total'] = df['floors_total'].astype('int')
df['parks_around3000'] = df['parks_around3000'].astype('int')
df['ponds_around3000'] = df['ponds_around3000'].astype('int')
```

`living_area`, `kitchen_area` и `days_exposition` заменим на медиану, так как количество пропусков невелико.

```python
living_area_median = df['living_area'].median()
df['living_area'].fillna(living_area_median, inplace=True)

kitchen_area_median = df['kitchen_area'].median()
df['kitchen_area'].fillna(kitchen_area_median, inplace=True)

days_exposition_median = df['days_exposition'].median()
df['days_exposition'].fillna(days_exposition_median, inplace=True)
df['days_exposition'] = df['days_exposition'].astype('int')
```
