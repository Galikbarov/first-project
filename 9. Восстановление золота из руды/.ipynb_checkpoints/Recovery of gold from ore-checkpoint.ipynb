–ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ. –í—ã—á–∏—Å–ª–∏–º –µ—ë –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ `rougher.output.recovery`. –ù–∞–π–¥—ë–º MAE –º–µ–∂–¥—É –Ω–∞—à–∏–º–∏ —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞.

* –° - –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏/–æ—á–∏—Ç—Å–∫–∏ - `rougher.output.concentrate_au`
* F - –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ/–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –¥–æ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏ - `rougher.input.feed_au`
* T - –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –æ—Ç–≤–∞–ª—å–Ω—ã—Ö —Ö–≤–æ—Å—Ç–∞—Ö –ø–æ—Å–ª–µ —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏ - `rougher.output.tail_au`

````python
c = train['rougher.output.concentrate_au']
f = train['rougher.input.feed_au']
t = train['rougher.output.tail_au']

recovery = ((c*(f-t))/(f*(c-t))) * 100
````

–ü–æ—Å—á–∏—Ç–∞–µ–º MAE:

````python
mean_absolute_error(train['rougher.output.recovery'], recovery)
````

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> –í–∏–∂—É, —Ö–æ—Ä–æ—à–æ.

–ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏, –Ω–µ–¥–æ—Å—Ç—É–ø–Ω—ã–µ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:

````python
inaccessible_signs = set(train.columns) - set(test.columns)
print(f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {len(inaccessible_signs)}')
inaccessible_signs
````

````python
rename_sings = {'final': '—Ñ–∏–Ω–∞–ª—å–Ω—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏',
                'primary_cleaner': '–ø–µ—Ä–≤–∏—á–Ω–∞—è –æ—á–∏—Å—Ç–∫–∞',
                'secondary_cleaner': '–≤—Ç–æ—Ä–∏—á–Ω–∞—è –æ—á–∏—Å—Ç–∫–∞',
                'input': '–ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Å—ã—Ä—å—è',
                'output': '–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –ø—Ä–æ–¥—É–∫—Ç–∞',
                'rougher': '—Ñ–ª–æ—Ç–∞—Ü–∏—è',
                'state': '–ø–∞—Ä–∞–º–µ—Ç—Ä—ã, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É—é—â–∏–µ —Ç–µ–∫—É—â–µ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ —ç—Ç–∞–ø–∞',
                'calculation': '—Ä–∞—Å—á—ë—Ç–Ω—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏'}
````

````python
separation_signs = {}

for sign in inaccessible_signs:
    list_sing = sign.split('.')
    step = list_sing[0]+'-'+list_sing[1]
    if step in separation_signs:
        separation_signs[step].append(list_sing[2])
    elif step not in separation_signs:
        separation_signs[step] = [list_sing[2]]
        
for key in separation_signs.keys():
    split_key = key.split('-')
    print()
    print(f'–°—Ç–∞–¥–∏—è {(rename_sings[split_key[0]])} —Å —Ñ—É–Ω–∫—Ü–∏–µ–π "{(rename_sings[split_key[1]])}":')
    for elem in separation_signs[key]:
        print(f' - {elem}')     
````

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è sMape:

````python
def get_smape(target, predict):
    smape = (1/len(target))*np.sum((np.abs(target - predict))/
                                   ( (np.abs(target)+np.abs(predict))/2))*100
    return smape
````

````python
smape_scorer = make_scorer(get_smape, greater_is_better = False)
````

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE.

````python
def get_final_smape(rougher, final):
    final = 0.25*rougher+0.75*final
    return final
````

–•–æ—Ç—è —Ç—ã –∏ –Ω–µ –∏—Å–ø–æ–ª—å–∑—É–µ—à—å —Ç—É—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—é –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –æ—Å—Ç–∞–≤–ª—é –¥–ª—è —Ç–µ–±—è –ø–æ–ª–µ–∑–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é. –î–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –∞ —Ç–∞–∫–∂–µ –¥–ª—è –º–æ–¥–µ–ª–µ–π, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã—Ö –Ω–∞ –¥–µ—Ä–µ–≤—å—è—Ö, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –ª–∏–±–æ –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö –ø–µ—Ä–µ–¥ –æ–±—É—á–µ–Ω–∏–µ–º –Ω–∏–∫–∞–∫ –Ω–µ –≤–ª–∏—è–µ—Ç –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (<a href = 'https://www.baeldung.com/cs/normalization-vs-standardization'>—Å—Å—ã–ª–∫–∞</a>)
        
–£ –º–æ–¥–µ–ª–µ–π –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–µ—Ä–µ–≤—å–≤ –ø—Ä–∏–Ω—Ü–∏–ø –ø—Ä–∏–Ω—è—Ç–∏—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–≤—è–∑–∞–Ω –Ω–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∞ –æ–Ω–∞ –ø—Ä–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏ –≤–æ–∑–¥–µ–π—Å—Ç–≤–∏—é –Ω–µ –ø–æ–¥–≤–µ—Ä–≥–∞–µ—Ç—Å—è. 
        
–° –ª–∏–Ω–µ–π–Ω—ã–º–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏—è–º–∏ —á—É—Ç—å –∏–Ω—Ç–µ—Ä–µ—Å–Ω–µ–µ: —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –Ω–∏–∫–∞–∫ —Å–µ–±—è –Ω–µ –ø—Ä–æ—è–≤–∏—Ç, –ø–æ—Å–∫–æ–ª—å–∫—É –≤—Å–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –º–æ–∂–Ω–æ –Ω–∏–≤–µ–ª–∏—Ä–æ–≤–∞—Ç—å –∑–∞ —Å—á–µ—Ç –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞:  

$y = \alpha + \beta X$ - –±–µ–∑ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏

$y = \alpha_{st} + \beta_{st} \frac{X-mean}{std} = (\alpha_{st} - \frac{\beta_{st}mean}{std}) + (\frac{\beta_{st}}{std}) X$ - —Å–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–µ–π

–ü—Ä–∏ —ç—Ç–æ–º $\alpha = (\alpha_{st} - \frac{\beta_{st}mean}{std})$ –∏ $\beta = (\frac{\beta_{st}}{std})$.

–ü–æ —Å—Å—ã–ª–∫–∞–º –º–æ–∂–Ω–æ —É–∑–Ω–∞—Ç—å, –∫–æ–≥–¥–∞ –≤—Å–µ-—Ç–∞–∫–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –∫—Ä–∞–π–Ω–µ –≤–∞–∂–Ω–∞: <a href = 'https://www.listendata.com/2017/04/how-to-standardize-variable-in-regression.html'>—Å—Å—ã–ª–∫–∞ 1</a>, <a href = 'https://www.statlect.com/fundamentals-of-statistics/linear-regression-with-standardized-variables'>—Å—Å—ã–ª–∫–∞ 2</a>.

–ü—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ –¥–∞–µ—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, —ç—Ç–æ —Ç–æ, —á—Ç–æ –º—ã –º–æ–∂–µ–º –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Å—É–¥–∏—Ç—å –æ –≤–∞–∂–Ω–æ—Å—Ç–∏ —Å–æ–æ—Ç–≤–µ—Å—Ç–≤—É—é—â–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–±–æ–ª—å—à–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç - –±–æ–ª—å—à–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç—å, –≤–∞–∂–Ω–æ—Å—Ç—å –Ω–µ–æ–±—Ö–æ–¥–∏–º–∞ –¥–ª—è —Ä–µ–≥—Ä–µ—Å—Å–∏–π —Å —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–µ–π). –ë–µ–∑ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏ —Ç–∞–∫–∞—è –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –Ω–µ–≤–æ–∑–º–æ–∂–Ω–∞, —Ç–∞–∫ –∫–∞–∫ –∫–∞–∂–¥—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤ —Å–µ–±—è –≤–∫–ª—é—á–∞–µ—Ç –µ—â–µ –∏ —ç–ª–µ–º–µ–Ω—Ç –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏ –º–∞—Å—à—Ç–∞–±–∞ —Å–≤–æ–µ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞. –ù–æ –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π —ç—Ç–æ –Ω–∏–∫–∞–∫ –Ω–µ –æ—Ç—Ä–∞–∂–∞–µ—Ç—Å—è.
        
–° –¥—Ä—É–≥–æ–π —Å—Ç–æ—Ä–æ–Ω—ã, –ø—Ä–∏–º–µ–Ω—è—è —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—é –∫ –ø—Ä–∏–∑–Ω–∞–∫–∞–º, –Ω—É–∂–Ω–æ –±—ã—Ç—å –æ—Å—Ç–æ—Ä–æ–∂–Ω—ã–º: –µ—Å–ª–∏ –Ω–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∫–æ–Ω–≤–µ–µ—Ä–Ω—É—é –æ–±—Ä–∞–±–æ—Ç–∫—É (<a href = 'https://stackoverflow.com/questions/51459406/how-to-apply-standardscaler-in-pipeline-in-scikit-learn-sklearn'>—á–µ—Ä–µ–∑ –ø–∞–π–ø–ª–∞–π–Ω</a>), —Ç–æ —É—Ç–µ—á–∫–∞ –¥–∞–Ω–Ω—ã—Ö –≤–Ω—É—Ç—Ä–∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –±—É–¥–µ—Ç –Ω–µ–∏–∑–±–µ–∂–Ω–∞. –í –ø–∞–π–ø–ª–∞–π–Ω–∞—Ö —Ä–µ–∞–ª–∏–∑—É–µ—Ç—Å—è —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è —Ç–µ–∫—É—â–µ–π —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏, –∏ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω–∏–µ –µ–µ –Ω–∞ —Ç–µ–∫—É—â—É—é –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é.

````python
%%time

state = RandomState(12345)

best_result_for_dt = 100
best_param_for_dt = 0
for n in range(1, 10, 1):
    model_dt_rougher = DecisionTreeRegressor(max_depth=n,
                                             random_state=state)
    smape_rougher = cross_val_score(model_dt_rougher,
                              features_for_train,
                              target_for_rougher_on_train,
                              scoring=smape_scorer,
                              cv=5)
    final_smape_rougher = smape_rougher.mean()

    model_dt_final = DecisionTreeRegressor(max_depth=n, random_state=state)
    smape_final = cross_val_score(model_dt_final, 
                                  features_for_train,
                                  target_for_final_on_train,
                                  scoring=smape_scorer,
                                  cv=5)
    final_smape_final = smape_final.mean()
    result_smape = get_final_smape(final_smape_rougher,
                                   final_smape_final)*(-1)
    
    if result_smape < best_result_for_dt:
        best_result_for_dt = result_smape
        best_param_for_dt = n
        
scores['DT'] = [best_result_for_dt, best_param_for_dt]
````

–ú–æ–¥–µ–ª—å RandomForestRegressor

````python
%%time

best_model_for_rf = 100
best_param_for_rf = 0

for n in range(10, 100, 30):
    for d in range(2,10,2):
        model_rf_rougher = RandomForestRegressor(n_estimators=n,
                                                 max_depth=d,
                                                 random_state=state)
        smape_rougher = cross_val_score(model_rf_rougher,
                                  features_for_train,
                                  target_for_rougher_on_train,
                                  scoring=smape_scorer,
                                  cv=5)
        final_smape_rougher = smape_rougher.mean()

        model_rf_final = RandomForestRegressor(n_estimators=n,
                                               max_depth=d,
                                               random_state=state)
        smape_final = cross_val_score(model_rf_final,
                                features_for_train,
                                target_for_final_on_train,
                                scoring=smape_scorer,
                                cv=5)
        final_smape_final = smape_final.mean()
        result_smape = get_final_smape(final_smape_rougher,
                                       final_smape_final)*(-1)
        
        if result_smape < best_model_for_rf:
            best_model_for_rf = result_smape
            best_param_for_rf = [n,d]
            
scores['RF'] = [best_model_for_rf, best_param_for_rf]
````
