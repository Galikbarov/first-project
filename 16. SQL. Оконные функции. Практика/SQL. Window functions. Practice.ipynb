# SQL. Оконные функции. Практика

**Описание:**

В нашем распоряжении база данных магазина, нам нужно проанализировать данные и написать запросы к базе. 

**Поставленные запросы:**

**Запрос №1.**

Рассчитаем сумму трат на привлечение пользователей с накоплением по месяцам c 2017 по 2018 год включительно.
Выгрузим два поля:
* месяц, приведённый к типу date;
* сумма трат на текущий месяц с накоплением.

В таблицу войдут данные о компаниях, в истории которых было больше шести важных этапов, а раунды финансирования проходили с 2012 по 2013 год включительно.

**Запрос №2.**

Посчитаем события с названием `view_item` по месяцам с накоплением. Рассчитаем количество событий только для тех пользователей, которые совершили хотя бы одну покупку.
Выгрузим поля: 
* месяц события, приведённый к типу date;
* количество событий за текущий месяц;
* количество событий за текущий месяц с накоплением.

**Запрос №3.**

Используя конструкцию WINDOW, рассчитаем суммарную стоимость и количество заказов с накоплением от месяца к месяцу.
Выгрузим поля:
* идентификатор заказа;
* месяц оформления заказа, приведённый к типу date;
* сумма заказа;
* количество заказов с накоплением;
* суммарная стоимость заказов с накоплением.

**Запрос №4.**

Напишем запрос, который выведет сумму трат на привлечение пользователей по месяцам, а также разницу в тратах между текущим и предыдущим месяцами. Разница должна показывать, на сколько траты текущего месяца отличаются от предыдущего. В случае, если данных по предыдущему месяцу нет, укажите ноль.
Выгрузим поля:
* месяц, приведённый к типу date;
* траты на привлечение пользователей в текущем месяце;
* разница в тратах между текущим и предыдущим месяцами.

**Запрос №5.**

Напишем запрос, который выведет сумму выручки по годам и разницу выручки между текущим и следующим годом. Разница должна показывать, на сколько выручка следующего года отличается от текущего. В случае, если данных по следующему году нет, укажите ноль.
Выгрузим поля:
* год, приведённый к типу date;
* выручка за текущий год;
* разница в выручке между текущим и следующим годом.
